```python
import os
from azure.cognitiveservices.vision.computervision import ComputerVisionClient
from azure.cognitiveservices.vision.computervision.models import OperationStatusCodes
from msrest.authentication import CognitiveServicesCredentials

def detect_text_with_azure(image_path, input_text):
    # Autentificare în Azure
    os.environ["VISION_KEY"] = "27d682fdc92c400a9aabd9d5aecde91e"
    os.environ["VISION_ENDPOINT"] = "https://ggg.cognitiveservices.azure.com/"

    # Accesarea cheii și a endpoint-ului din variabilele de mediu
    subscription_key = os.environ["VISION_KEY"]
    endpoint = os.environ["VISION_ENDPOINT"]

    # Crearea clientului Computer Vision
    computervision_client = ComputerVisionClient(endpoint, CognitiveServicesCredentials(subscription_key))

    # Detectare text în imagine
    with open(image_path, "rb") as img:
        result = computervision_client.read_in_stream(image=img, raw=True)

    # Verificare dacă operația a fost finalizată cu succes
    operation_location = result.headers["Operation-Location"]
    operation_id = operation_location.split("/")[-1]
    operation_status = None

    while operation_status != OperationStatusCodes.succeeded:
        result_status = computervision_client.get_read_result(operation_id)
        operation_status = result_status.status

    # Obținerea rezultatelor analizei
    read_result = computervision_client.get_read_result(operation_id)

    # Extrage textul detectat
    detected_text = ""
    for text_result in read_result.analyze_result.read_results:
        for line in text_result.lines:
            detected_text += line.text + " "

    # Afișare rezultat
    print("Textul detectat în imagine:")
    print(detected_text)

    # Verificare corectitudine text
    input_words = input_text.split()
    detected_words = detected_text.split()
    if all(word in detected_words for word in input_words):
        return True
    else:
        return False

# Calea către imaginea din Google Drive
image_path = "/content/drive/My Drive/test2.jpeg"

# Textul de intrare pentru verificare
input_text = "Succes in rezolvarea tEMELOR la LABORAtoarele de Inteligenta Artificiala!"

# Detectarea textului în imagine utilizând Azure Computer Vision și verificarea corectitudinii
if detect_text_with_azure(image_path, input_text):
    print("Textul a fost localizat corect în imagine.")
else:
    print("Textul nu a fost localizat corect în imagine.")
```
