```python
def hamming_distance(str1, str2):
    """
    Calculează distanța Hamming între două șiruri de caractere de aceeași lungime.
    """
    if len(str1) != len(str2):
        raise ValueError("Lungimea șirurilor trebuie să fie aceeași.")
    return sum(ch1 != ch2 for ch1, ch2 in zip(str1, str2))

def hamming_word_distance(text1, text2):
    """
    Calculează distanța Hamming între două texte la nivel de cuvânt.
    """
    words1 = text1.split()
    words2 = text2.split()
    if len(words1) != len(words2):
        raise ValueError("Numărul de cuvinte trebuie să fie același în ambele texte.")
    return sum(word1 != word2 for word1, word2 in zip(words1, words2))

# Textul recunoscut
real_text = "Lucces in rezolvarea TEMELOR la LABORA toate de Inteligenta Artificiala!"
recognized_text = "Lucces in resolvarea TEMELOR la LABORA toarele de Inteligenta Artificiala!"

# Eliminarea spațiilor suplimentare
real_text = " ".join(real_text.split())
recognized_text = " ".join(recognized_text.split())

print("Lungimea textului recunoscut:", len(recognized_text))
print("Lungimea textului real:", len(real_text))

# Calcularea distanței Hamming la nivel de caracter
character_distance = hamming_distance(recognized_text, real_text)
print("Distanța Hamming la nivel de caracter:", character_distance)

# Calcularea distanței Hamming la nivel de cuvânt
word_distance = hamming_word_distance(recognized_text, real_text)
print("Distanța Hamming la nivel de cuvânt:", word_distance)
```
