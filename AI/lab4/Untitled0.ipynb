```python
from azure.cognitiveservices.vision.computervision import ComputerVisionClient
from msrest.authentication import CognitiveServicesCredentials
import os
from azure.cognitiveservices.vision.customvision.prediction import CustomVisionPredictionClient
```

```python
from google.colab import drive

# Montați Google Drive
drive.mount('/content/drive')
# Setarea cheii și a endpoint-ului în variabile de mediu
os.environ["VISION_KEY"] = "27d682fdc92c400a9aabd9d5aecde91e"
os.environ["VISION_ENDPOINT"] = "https://ggg.cognitiveservices.azure.com/"

# Accesarea cheii și a endpoint-ului din variabilele de mediu
subscription_key = os.environ["VISION_KEY"]
endpoint = os.environ["VISION_ENDPOINT"]

# Crearea clientului Computer Vision
computervision_client = ComputerVisionClient(endpoint, CognitiveServicesCredentials(subscription_key))

# Inițializarea clientului pentru predicții
credentials = CognitiveServicesCredentials(subscription_key)
predictor = CustomVisionPredictionClient(endpoint, credentials)  # Aici am înlocuit 'ENDPOINT' cu 'endpoint'

# Funcție pentru clasificarea imaginilor
def classify_image(image_path):
    with open(image_path, "rb") as image_file:
        image_data = image_file.read()

    # Trimiterea imaginii la modelul de clasificare
    results = predictor.classify_image("3b3352f6-a029-42f7-a8ac-14a83bd7da97", "AI_GaldeanuNicole", image_data)

    # Procesarea rezultatelor
    for prediction in results.predictions:
        print(f"{prediction.tag_name}: {prediction.probability:.2f}")

# Testarea algoritmului de clasificare
image_path = '/content/drive/MyDrive/images/animals.png'
classify_image(image_path)

```
