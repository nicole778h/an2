```python
import matplotlib.pyplot as plt
import pandas as pd
import requests

# Descărcarea datelor de la link
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
response = requests.get(url)

# Scrierea datelor într-un fișier CSV local
with open("breast_cancer_dataset.csv", "wb") as f:
    f.write(response.content)

# Citirea datelor și specificarea numelor coloanelor
column_names = ['ID', 'Diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
                'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave_points_mean',
                'symmetry_mean', 'fractal_dimension_mean', 'radius_se', 'texture_se', 'perimeter_se',
                'area_se', 'smoothness_se', 'compactness_se', 'concavity_se', 'concave_points_se',
                'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst', 'perimeter_worst',
                'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave_points_worst',
                'symmetry_worst', 'fractal_dimension_worst']
data = pd.read_csv("breast_cancer_dataset.csv", header=None, names=column_names)

# Funcția pentru afișarea histogramei pentru variabila dată
def plot_histogram(data, xlabel, title):
    plt.hist(data, bins=20, color='skyblue', edgecolor='black')
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Afisarea histogramei pentru caracteristica "texture_mean"
plot_histogram(data['texture_mean'], 'Texture Mean', 'Distribution of Texture Mean')

# Afisarea histogramei pentru caracteristica "radius_mean"
plot_histogram(data['radius_mean'], 'Radius Mean', 'Distribution of Radius Mean')
```
